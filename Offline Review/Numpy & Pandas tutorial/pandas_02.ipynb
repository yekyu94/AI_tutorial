```python
# 간단히 Columns을 추가할 수 잇습니다.
df["sum"] = df["Jan"]+df["Feb"]+df["Mar"]
df.tail()
```

```python
# 엑셀에서 사용하는 일부 기능을 동일하게 구현할 수 있습니다.
df["Jan"].sum(), df["Jan"].mean(),df["Jan"].min(),df["Jan"].max()
```

```python
# 특정 columns들의 값에대한 정보만도 확인이 가능합니다.
df[['Jan','Feb']].describe()
```

```python
# unique()명령은 특정 column의 요소들의 중복을 제거한 후 출력합니다.
# 이 명령을 통해 아래와 같이 nan값과 같은 데이터의 결측치가 존재하는지 여부를 확인할 수 있습니다.
df['Jan'].unique()
```

```python
# groupby 명령은 데이터베이스에서 사용하는 SQL명령과 동일한 기능을 합니다.
# DataFrame[[출력할 column명]].groupby(그룹화할 column).(그룹화된 녀석들에게 할 명령)
df_sub=df[['state',"Jan","Feb","Mar","sum"]].groupby('state').sum()
df_sub
# 아래 출력결과를 확인해보면, state명이 같은 경우 각 column(=lowa, Rhodelsland)의 값을 합친 결과가 나온것을 볼 수 있습니다.
# 0이 출력된 것은 NaN값이 계산에 포함되었기 때문입니다.
```
